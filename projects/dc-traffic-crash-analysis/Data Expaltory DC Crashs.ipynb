```python
df2=df[['REPORTDATE','MEASURE','FROMDATE','ADDRESS','WARD','MAR_ADDRESS', 'MAR_SCORE', 'MAJORINJURIES_BICYCLIST',
       'MINORINJURIES_BICYCLIST', 'UNKNOWNINJURIES_BICYCLIST',
       'FATAL_BICYCLIST', 'MAJORINJURIES_DRIVER', 'MINORINJURIES_DRIVER',
       'UNKNOWNINJURIES_DRIVER', 'FATAL_DRIVER', 'MAJORINJURIES_PEDESTRIAN',
       'MINORINJURIES_PEDESTRIAN', 'UNKNOWNINJURIES_PEDESTRIAN',
       'FATAL_PEDESTRIAN', 'TOTAL_VEHICLES', 'TOTAL_BICYCLES',
       'TOTAL_PEDESTRIANS', 'PEDESTRIANSIMPAIRED', 'BICYCLISTSIMPAIRED',
       'DRIVERSIMPAIRED', 'TOTAL_TAXIS', 'TOTAL_GOVERNMENT',
       'SPEEDING_INVOLVED', 'NEARESTINTROUTEID', 'NEARESTINTSTREETNAME']]
```

```python
df2[['Year','Month','Day']]= df2['REPORTDATE'].str.split('/',expand=True)
```

```python
df2['Day']= df2['Day'].str.split(' ').str[0]
```

```python
Summery = df2.groupby('Year')[['MAJORINJURIES_BICYCLIST']].sum().reset_index()
```

```python
print(Summery)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Your Summery DataFrame already exists:
# Summery = df2.groupby('Year')[['MAJORINJURIES_BICYCLIST']].sum().reset_index()

plt.figure(figsize=(14,7))  # 🔑 make the chart large

# Create bar chart
plt.bar(Summery["Year"], Summery["MAJORINJURIES_BICYCLIST"], color="skyblue")

# Add labels and title
plt.xlabel("Year", fontsize=14)
plt.ylabel("Major Injuries (Bicyclists)", fontsize=14)
plt.title("Major Injuries (Bicyclists) by Year", fontsize=16)

# Rotate year labels for readability
plt.xticks(rotation=45, ha="right", fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis with commas
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# Add grid for clarity
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Your Summery1 DataFrame already exists:
# Summery1 = df2.groupby('Year')[['MINORINJURIES_DRIVER']].sum().reset_index()

plt.figure(figsize=(14,7))  # 🔑 big size to match your bar chart

# Create line plot
plt.plot(Summery1["Year"], Summery1["MINORINJURIES_DRIVER"], 
         color="red", marker="o", linewidth=2)

# Labels and title
plt.xlabel("Year", fontsize=14)
plt.ylabel("Minor Injuries (Drivers)", fontsize=14)
plt.title("Minor Injuries (Drivers) by Year", fontsize=16)

# Rotate x-axis labels
plt.xticks(rotation=45, ha="right", fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis with commas
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# Add grid for clarity
plt.grid(True, linestyle="--", alpha=0.7)

plt.show()
```

```python
Summery2 = df2.groupby('Year')[['FATAL_DRIVER']].sum().reset_index()
Summery2
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Your Summery2 DataFrame already exists:
# Summery2 = df2.groupby('Year')[['FATAL_DRIVER']].sum().reset_index()

plt.figure(figsize=(14,7))  # 🔑 same big size

# Create bar chart with different colors
colors = plt.cm.tab10.colors  # a set of distinct colors
plt.bar(Summery2["Year"], Summery2["FATAL_DRIVER"], 
        color=[colors[i % len(colors)] for i in range(len(Summery2))])

# Labels and title
plt.xlabel("Year", fontsize=14)
plt.ylabel("Fatalities (Drivers)", fontsize=14)
plt.title("Fatal Driver Injuries by Year", fontsize=16)

# Rotate x-axis labels
plt.xticks(rotation=45, ha="right", fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis with commas
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# Add grid lines
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```
